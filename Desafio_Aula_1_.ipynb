## Caderno de exploração dos dados do DATASUS 2008 para frente

```python
import pandas as pd
```

```python
dados = pd.read_csv("/content/A103939189_28_143_208.csv", encoding="ISO-8859-1",
                    skiprows=3, sep=";", skipfooter=12,
                    thousands=".", decimal=",")
dados.head()
```

```python
dados.plot(x="Unidade da Federação", y="2008/Ago")
```

```python
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por unidade de federação")
plt.show()
```

# Desafio 01: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.

```python
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor do processamento de procedimentos hospitalares do SUS em Agosto de 2008 por Unidade da Federação")
plt.show()
```
